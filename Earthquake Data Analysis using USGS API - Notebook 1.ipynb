This assignment collects earthquake data that is detailed for various locations around the globe, obtained from the U.S. Geological Survey Earthquake API. Ordinarily, earthquake data forms a key aspect in the analysis and study of seismic activities, assessment of the risks arising from earthquakes, and enhancement of preparedness and response actions.

```python
import requests
import pandas as pd
url = "https://earthquake.usgs.gov/fdsnws/event/1/query"
response = requests.get(url, params={"format": "geojson"})
data = response.json()
features = data['features']
df = pd.json_normalize(features)
df.to_csv("earthquake.csv")
```
